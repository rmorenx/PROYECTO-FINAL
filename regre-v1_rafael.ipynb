```python
import pandas as pd
music= pd.read_csv('music.csv')
music
```

```python
missing_values = music.isnull().sum()
print(missing_values)
```

```python
columns2drop= ["released_year", 
               "released_month", 
               "released_day",
               "cover_url",
               "artist_count",
               "in_deezer_playlists",
               "in_spotify_playlists",
               "in_apple_playlists",
               "in_spotify_charts",
               "in_apple_charts",
               "in_deezer_charts",
               "in_shazam_charts"

              ]

music_cleaned = music.drop(columns=columns2drop, errors='ignore')
music_cleaned.to_csv("music_clean.csv", index=False)

music_cleaned_sorted = music_cleaned.sort_values(by="streams", ascending=False)
music_cleaned_sorted.to_csv("music_sorted", index= False)

music_cleaned_sorted['streams'] = pd.to_numeric(music_cleaned_sorted['streams'], errors='coerce')
music_cleaned_sorted_numeric = music_cleaned_sorted.dropna(subset=['streams'])
music_cleaned_sorted_numeric.to_csv("music_cleaned_sorted_numeric.csv", index=False)

```
